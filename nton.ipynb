```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("synthetic_laptop_data.csv")
print(df.head())
print("\n")
# Features and target
X = df[['RAM_GB', 'GPU_Score', 'Processor_GHz']]
y = df['Type']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Plotting the feature coefficients
features = ['RAM_GB', 'GPU_Score', 'Processor_GHz']
coefficients = model.coef_[0]

plt.bar(features, coefficients, color='skyblue')
plt.title('Feature Coefficients (Weights)')
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.show()

# Custom input prediction
custom_laptop = pd.DataFrame([[6, 5000,10]], columns=['RAM_GB', 'GPU_Score', 'Processor_GHz'])
prediction = model.predict(custom_laptop)
print("Recommended For:", "Gaming" if prediction[0] == 1 else "General Use")
```
